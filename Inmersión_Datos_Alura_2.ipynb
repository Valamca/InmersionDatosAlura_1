```python
plt.figure(figsize = (10,8)) ##Nuevo lienzo
colores = {"Apartamento":"cyan","Casa":"yellow","Finca":"red"}
grafica = sns.histplot(data = dfdatos, x="Precio_millon", hue = "Tipo",multiple="dodge",palette = colores,edgecolor = "gray")#,kde = True para la linea de tendencia
grafica.set_title("Distribución de valores inmuebles Bogota por Tipo") #ajustamos el nombre del Histograma
plt.xlabel("Precio en millones")
plt.ylabel("Número de inmuebles disponibles")
plt.xlim(500,3000) #Limites eje X
plt.ylim(0,300) #Limites eje Y
plt.show()
```

Cómo se puede observar la disponibilidad mayoritaria de inmuebles a menor precio son los apartamentos, mientrás que entre mas se aumenta el rango de precios, comienzan a aparecer las fincas, a su vez las casas son estables tanto sus precios como disponibilidad en rangos inferiores a 1200 millones de pesos colombianos.

**¿Cuál es el precio del m2 de área por Barrio?**

```python
inmuebles[["Área","Precio_millon"]] #observamos información para enteder el contexto de la pregunta y nuestra información directa
```

```python
inmuebles["Precio_m^2_area"]= inmuebles["Precio_millon"]/inmuebles["Área"]  # Campo que necesitamos para nuestros calculos, el precio/area para saber el precio por m2
```

*Es posible realizar este calculo debido a que los campos son INT y Float respectivamente*

```python
inmuebles.info() #Aquí podemos ver los tipos de objetos de cada columna y sus valores indexados
```

```python
dfPrecioM2 = inmuebles[["Barrio","Precio_m^2_area"]] ##Obtenemos las columnas necesarias para responder a la pregunta
dfPrecioM2 = dfPrecioM2.groupby(by="Barrio").mean() #Hacemos el filtro por Barrios y la obtención del promedio del precio del area por cada uno
dfValorM2 = dfPrecioM2.sort_values(by = "Precio_m^2_area",ascending = True).reset_index() #Ordenamos lo valores para mejor manejo, y aplicamos un formato df con index
dfValorM2 = dfValorM2.tail(15) #Imprimo los primeros 15 datos para entender su tendencía
dfValorM2
```

*Graficamos para tener una clara visión de nuestros resultados*

```python
dfPrecioM21 = inmuebles[["Barrio","Precio_m^2_area"]] ##Obtenemos las columnas necesarias para responder a la pregunta
dfPrecioM21 = dfPrecioM2.groupby(by="Barrio").mean() #Hacemos el filtro por Barrios y la obtención del promedio del precio del area por cada uno
dfValorM21 = dfPrecioM2.sort_values(by = "Precio_m^2_area",ascending = True).reset_index() #Ordenamos lo valores para mejor manejo, y aplicamos un formato df con index
dfValorM21 = dfValorM2.head(15) #Imprimo los primeros 15 datos para entender su tendencía
dfValorM21
```

```python
plt.figure(figsize = (10,8)) ##Nuevo lienzo
fig, ax = plt.subplots()
ax.stem(dfValorM2["Barrio"],dfValorM2["Precio_m^2_area"],orientation = "horizontal",linefmt = 'k--', basefmt = 'k:',markerfmt = 'or' )
plt.xlabel("Precio cada m^2 de area en millones")
plt.ylabel("Barrio")
plt.show()
```

```python
plt.figure(figsize = (10,8)) ##Nuevo lienzo
fig, ax = plt.subplots()
ax.stem(dfValorM21["Barrio"],dfValorM21["Precio_m^2_area"],orientation = "horizontal",linefmt = 'k--', basefmt = 'k:',markerfmt = 'or' )
plt.xlabel("Precio cada m^2 de area en millones")
plt.ylabel("Barrio")
plt.show()
```

Cómo podemos observar en los gráficos, el metro cuadrado es mas caro en el Barrio llamado "El Virrey" haciendo honor a su nombre, mientrás que el valor mas bajo por metro cuadrado se encuentra en el Barrio llamado "Bosa Piamonte", con una diferencia de casí el 90% en el precio entre ambos extremos del análisis, lo cuál nos habla una zona con valor de inmuebles muy bajos.
