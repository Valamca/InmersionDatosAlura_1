**En la base anterior la columna "VALOR" es un String así que necesitamos convertirlo a númerico.**

Para esto utilizaremos la función "Split" después del comando .str por tratarse de un String

```python
valor = inmuebles.Valor.str.split(expand = True) ##Devuelve cada elemento de la columna cómo una lista de valores serparados en un Data Frame
inmuebles["Moneda"] = valor[0] #Creamos una nueva columna con el valor creado en el dataframe
inmuebles["Precio"] = valor[1] #creamos otra columna con el siguiente valor creado en el dataframa
inmuebles.head()
```

Cómo podemos observar en el DataFrame, aún tenemos que eliminar el . de los valores en la columna "Precio":
lo cual se llevará a cabo con la función "Replace" para sustituir un "." con por un lugar vacio "".

```python
inmuebles["Precio"] = inmuebles["Precio"].str.replace(".","",regex = True) #Eliminamos el punto decimal de todos los valores de la columna valor
inmuebles[["Barrio","Precio"]] #se obtiene la información a trabajar
```

**IMPORTANTE** <br> Una vez que ya hayamos quitado todo caracter que nos impida poder convertir el texto a númerico, hacemos uso de la función "Astype" para convertir desde un STRING a un float.

```python
inmuebles["Precio_millon"] = inmuebles.Precio.astype("float")/1000000 #Convertimos los valores en un tipo Float y lo guardamos en una nueva columna
inmuebles.info() ##Revisamos si la nueva columna es del tipo especificado
```

### ***¿Cuál es el precio del m2 de área por Barrio?***

## Comenzamos con los procesos de Machine Learning <br>
Importamos las librerias que contienen los métodos necesarios para realizar nuestros calculos y entrenamientos

```python
from sklearn.linear_model import LinearRegression #Importamos la librería con los métodos para hacer las regresiones
from sklearn.model_selection import train_test_split #Importamos para entrenar nuestros modelos generados

X = datos_ml[["COD_UPZ_GRUPO"]] ##Ajustamos nuestros ejes con nuestra variable X
y = datos_ml["Precio_millon"] ## Ajustamos nuestros ejes con nuestra variable y

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=99)

##Explico la razón de la formula de arriba a continuación:
```

#### La Expresión: <br>
>   **train_test_split(X,y,test_size=0.25,random_state=99)** <br>
<br>
Obtiene sus datos de las variables "X" y "y" que se ajusataron antes de la expresión y su funcionamiento es el siguiente: <br>
<br>
Extrae el 25% de los datos totales en "X" y "y" y los almacena en las variables con terminación "_test", seleccionandolos de manera aleatoria. <br>
Lo que da como resultado 2 variables X(X_train y X_test), una con el 75% y la otra con el 25% restante de datos de su respetivo eje, en este caso del eje X. Se llevan a cabo el mismo procedimiento pero ahora con el eje Y, resultado con las dos variables y(y_train y y_test) con la misma proporción de datos, para dar un total de 4 variables generadas con la expresión **train_test_split** <br>
<br>
_Nota_: Utilizamos el atributo "random_state" para controlar la aleatoriedad del modelo, es decir que si otra persona usa el mismo valor semilla, en este caso "99" obtendrá los mismo resultados "Aleatorios" en cada variable generada.

```python
X_train.shape #Observamos la cantidad de datos almacenados en la variable con el 75% de los datos X
```

```python
X_test.shape #Observamos la cantidad de datos almacenados en la variable con el 25 de los datos X
```

```python
y_train.shape #Observamos la cantidad de datos almacenados en la variable con el 75% de los datos y
```

```python
y_test.shape #Observamos la cantidad de datos almacenados en la variable con el 25 de los datos y
```

### Ajuste de modelo de Machine Learning 

>**En este caso el modelo seleccionado para ajustar fue uno de Regresión lineal** <br>
Este tipo de modelo se enfoca en ajustar los datos en un recta pendiente con ecuación simple: **_y = mx+b_**

```python
modelo = LinearRegression() #Instanciamos un modelo de regresión lineal en la variable "modelo" para mas comodidad

modelo.fit(X_train,y_train) # Ajustamos el modelo con el atributo "fit", y variables con el 75% de información para entrenarlo
```

```python
y_predict_test = modelo.predict(X_test) #Aquí utilizamos el atributo "Predict" del modelo para obtener valores "y" con base
                                       # en los datos almacenados en las variables "_test" con el 25% restante de los datos
```

**Medición de eficacia del modelo**

```python
from sklearn.metrics import mean_absolute_error,r2_score ## Librerias para medir la eficiencia del modelo ajustado

baseline_mae = mean_absolute_error(y_test,y_predict_test) ## Rango de variación de errores (Límite Superior e Inferior)

baseline_r2 = r2_score(y_test,y_predict_test) ## Coeficiente de Determinación, para saber el % de acierto del modelo

print("Datos de Eficiencia: "+ "R2: "+ str(baseline_r2), "MAE: "+ str(baseline_mae)) ## Impresión de datos de EFICIENCIA
```

**Resultados**<br>
Como podemos observar en los resultados obtenidos por el modelo ajustado con solo 1 Variable para el eje "X", <br>
nuestro modelo solo puede predecir el 3% (R2 = 0.0311) de los datos lo cuál lo convierte en un modelo terrible, en el cual los valores que se obtengan de las estimaciones de precios tendran un rango de error de +-187.3 millones. <br>

Para comprender mejor lo anterior supongamos el siguiente ejemplo con el mismo modelo:<br>

**X = modelo[[9]]** <----UPZ con número de identificación "9"<br> 

Sí corremos este ejemplo, el valor obtenido sólo tendrá un 3% de probabilidad de ser correcto y el mismo tendrá un rango de variabilidad de +-187.3 millones para estimar el precio de un inmueble sí solo utilizamos el dato "COD_UPZ_GRUPO", como variable de correlación ó variable predictoria(independiente). <br>
<br>
Para evitar estos valores tan pobres, se opta por usar mas variables predictorias y generar un mejor resultado de determinación cómo en el siguiente ejemplo:  <br>




**Modelo con mas de 1 variable predictoria**

```python
##Aquí añadimos más variables al eje X para ajustar mejor nuestro modelo haciéndolo más significativo.
X = datos_ml[['COD_UPZ_GRUPO','Habitaciones','Baños','CONJUNTO_CERRADO','SALARIO_ANUAL_MI','TIENE_ESCRITURA']] 
Y = datos_ml["Precio_millon"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 99) ##Generamos variables necesarias

modelo_1 = LinearRegression() #Instaciamos nuestro modelo
modelo_1.fit(X_train, y_train) #Ajustamos nuestro modelo con nuestros datos con el 75% de información

y_predict_test = modelo_1.predict(X_test) #Generamos las pruebas con nuestros datos con el 25% de información
y_predict_train = modelo_1.predict(X_train) #Generamos mas pruebas ahora con nuestros datos de entrenamiento con el 75%

#Eficacioa de modelo con datos de prueba(El 25% de los datos)
mae_test = mean_absolute_error(y_test, y_predict_test) #Rango absoluto de error (Intervalo de confianza)
r2_test = r2_score(y_test, y_predict_test) #Coeficiente de Determinación (% de exito de nuestro modelo)

#Eficacia del modelo con datos de entrenamiento(El 75% de los datos)
mae_train = mean_absolute_error(y_train, y_predict_train) #Rango absoluto de error (Intervalo de confianza)
r2_train = r2_score(y_train, y_predict_train) #Coeficiente de Determinación (% de exito de nuestro modelo)

#Impresión de resultados en pantalla
print("Datos Test " + "MAE: "+str(mae_test),"R2: " + str(r2_test)) 
print("Datos Train " + "MAE: "+str(mae_train),"R2: "+ str(r2_train))
```
