**Top 10 barrios con mayor promedio de Area**

*Consultamos las áreas promedio, agrupadas por cada barrio*

```python

dfBarrios = inmuebles[["Barrio","Area"]] #Extraemos los datos que necesitamos del df
promedioPorBarrio = dfBarrios.groupby("Barrio").mean().reset_index() #Obtenemos sus medias agrupadas por cada barrio
promedioPorBarrio #Imprimimos los valores resultantes
```

*Ordenamos los datos y obtenemos el top 10*

```python
AreaPromedioBarrios = promedioPorBarrio.sort_values(by ="Area",ascending = False) #Ordenamos los valores de manera DESCENDENTE
Top10Barrios = AreaPromedioBarrios.head(10) #Obtenemos el top 10 de barrios por su área en venta.
Top10Barrios
```

*Graficamos el Top 10 de Barrios por Area promedio en venta*

```python
import matplotlib.pyplot as plt
#Esta libreria nos ayudara a realizar los gráficos
```

```python
Top10Barrios.plot.bar(x="Barrio",y="Area",color = "red")
#Modificando la gráfica para mejor comprensión
plt.xlabel("Barrios")
plt.ylabel("Área promedio m^2")
plt.title("Top 10 Barrios por Área promedio")

```

**¿Cuá es la mediana de baños que ofrecen los inmuebles por UPZ?**

```python
BañosPorUPZ = inmuebles[["UPZ","Banos"]] ##Obtenemos los datos a utilizar
BañosPorUPZ #listamos datos para comprender el contexto de la información
```

```python
dfBaños = BañosPorUPZ.groupby("UPZ").median().reset_index() #Obtenemos los datos filtrados
dfBaños ##Obtenemos la mediana de baños  por UPZ
```

```python
df = dfBaños.drop_duplicates() ##Eliminamos Duplicados
df = df.sort_values("Banos",ascending = False) ##Ordenamos valor en orden decendente
df ##Listamos los datos ya limpios y ordenados
```

Graficamos para reponder de manera gráfica la pregunta realizada

```python
df.head(10).plot.bar(x="UPZ",y="Banos",color = "green") ##Graficamos
plt.xlabel("UPZ") #Ajustamos eje X
plt.ylabel("Número de Baños") ##Ajustamos eje Y
```

**¿Cuál es el número promedio de habitaciones para cada tipo de inmueble?, y su numero total de tipos.**


```python
NHabitaciones = inmuebles[["Habitaciones","Tipo"]] ##Obtenemos información necesaria
NHabitaciones = NHabitaciones.groupby("Tipo").mean().reset_index() ##Ordenamos la información para responder las preguntas necesarias
NHabitaciones.round() ##Listamos información limpia y respondemos la primer pregunta
```
